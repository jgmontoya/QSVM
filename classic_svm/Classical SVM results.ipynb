# Best classical SVM results

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from AuxFunctions import plot_confusion_matrix
```

```python
df_train = pd.read_csv('../dataset/Iris_training.csv', index_col=0)
df_test = pd.read_csv('../dataset/Iris_testing.csv', index_col=0)
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
clf = SVC(C=1, gamma=0.3877551020408163, kernel="rbf").fit(df_train[features], df_train['Species'])
```

```python
train_pred = clf.predict(df_train[features])
test_pred = clf.predict(df_test[features])
```

```python
print(classification_report(df_train['Species'], train_pred))
```

```python
# Plot non-normalized confusion matrix
plot_confusion_matrix(df_train['Species'], train_pred, classes=np.array(class_names),
                      title='Training Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(df_train['Species'], train_pred, classes=class_names, normalize=True,
                      title='Training Normalized confusion matrix')

plt.show()
```

```python
print(classification_report(df_test['Species'], test_pred))
```

```python
# Plot non-normalized confusion matrix
plot_confusion_matrix(df_test['Species'], test_pred, classes=np.array(class_names),
                      title='Testing Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(df_test['Species'], test_pred, classes=class_names, normalize=True,
                      title='Testing Normalized confusion matrix')

plt.show()
```
